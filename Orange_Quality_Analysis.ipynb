#1.Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?

```python
data=pd.read_csv(r"Orange Quality Data.csv",usecols=['Size (cm)','Weight (g)'])
```

```python
total=data.groupby('Size (cm)').corr()
total.sort_values(by='Size (cm)',ascending=False)
```

```python
plt.figure(figsize=(18, 9))
sns.barplot(x='Size (cm)', y='Weight (g)', data=data)
plt.title('Correlation between Size and Weight of Oranges')
plt.xlabel('Size (cm)')
plt.ylabel('Weight (g)')
plt.show()
```

#3.How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))

#4.Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?

```python
df = data[["Blemishes (Y/N)","Ripeness (1-5)"]]
df1 = df
df1
```

```python
plt.figure(figsize = (16,4))
sns.barplot(data = df1,x = "Blemishes (Y/N)",y = "Ripeness (1-5)",hue = "Ripeness (1-5)") # Changed sn to sns
plt.title("Blemishes tends to Lower Ripeness")
plt.show()
```

```python
#b)"Blemishes (Y/N)","Softness (1-5)"
df = data[["Blemishes (Y/N)","Softness (1-5)"]]
df
```

```python
plt.figure(figsize = (15,4))
sns.barplot(data = df,x = "Blemishes (Y/N)",y = "Softness (1-5)",hue = "Softness (1-5)") # Changed sn to sns
plt.title("Blemishes tends to Soften Levels")
plt.show()
```

```python
sns.pairplot(data[["Blemishes (Y/N)","Ripeness (1-5)","Softness (1-5)"]],hue = "Blemishes (Y/N)") # Changed sn to sns
plt.show()
```
